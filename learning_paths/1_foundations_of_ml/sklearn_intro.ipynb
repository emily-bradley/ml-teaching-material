```python
numeric_cols = ['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6']

ordinal_cols = ['Education']

ohe_cols = ['Hobby', 'VideoGame']
```

```python
# Build the pipeline
numeric_transformer = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='mean')),
    ('scale', MinMaxScaler())
])

ordinal_transformer = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='constant')),
    ('encode', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)),
])

ohe_transformer = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='constant')),
    ('encode', OneHotEncoder(handle_unknown='ignore')),
])

preprocessor = ColumnTransformer(
    transformers=[
        ('numeric', numeric_transformer, numeric_cols),
        ('ordinal', ordinal_transformer, ordinal_cols),
        ('ohe', ohe_transformer, ohe_cols),
    ],
)

pipeline = Pipeline(steps=[
    ('preprocessing', preprocessor),
    ('linear_regression', LinearRegression())
])
```

```python
pipeline.fit(X_diabetes_train, y_diabetes_train)
```

## Putting it All Together
* Load the data
* Train/Test Split
* Transform in a Pipeline
* CV Grid Search to select best parameters
* Evaluate best_estimator on hold out set
* Train over entire dataset

Load the data and Train/Test Split

```python
bc = load_breast_cancer(as_frame=True)
X = bc.data
y = bc.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y,random_state=seed)
```

```python
pd.DataFrame(y_train).value_counts(normalize=True)
```

Create the Pipeline and Grid Search

```python
X_train.head()
```

```python
# Build the pipeline
numeric_cols = X_train.columns

numeric_transformer = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='median')),
    ('scale', MinMaxScaler())
])

preprocessor = ColumnTransformer(
    transformers=[
        ('numeric', numeric_transformer, numeric_cols),
    ],
)

pipeline = Pipeline(steps=[
    ('preprocessing', preprocessor),
    ('random_forest', RandomForestClassifier())
])
```

```python
# Build the grid search
rf_param_grid = [
  {'random_forest__n_estimators': [100, 150, 250],
   'random_forest__max_depth': [7, 20, 50, None],
   'random_forest__min_samples_split': [10],
   'random_forest__min_samples_leaf': [5, 20],
   'random_forest__random_state': [seed]},
 ]

grid_search = GridSearchCV(estimator=pipeline,
                           param_grid=rf_param_grid,
                           scoring=['f1', 'accuracy', 'neg_log_loss', 'roc_auc', 'jaccard'],
                           refit='f1',
                           cv=5,
                           n_jobs=8,
                           verbose=1,
                           error_score="raise")

grid_search.fit(X_train, np.ravel(y_train))
```

Evaluate Best_Estimator on Hold-Out Set

```python
best_estimator = grid_search.best_estimator_
```

```python
from sklearn.metrics import roc_auc_score, jaccard_score, accuracy_score, f1_score, precision_score, recall_score

preds = best_estimator.predict(X_test)

print(f'Validation roc_auc_score:  {round(roc_auc_score(np.ravel(y_test), best_estimator.predict_proba(X_test)[:, 1]), 3)}')
print(f'Validation accuracy_score:  {round(accuracy_score(np.ravel(y_test), preds), 3)}')
print(f'Validation jaccard_score:  {round(jaccard_score(np.ravel(y_test), preds), 3)}')
print(f'Validation f1_score:  {round(f1_score(np.ravel(y_test), preds), 3)}')
print(f'Validation precision_score:  {round(precision_score(np.ravel(y_test), preds), 3)}')
print(f'Validation recall_score:  {round(recall_score(np.ravel(y_test), preds), 3)}')
```

### Transformers

```python
from sklearn.base import BaseEstimator, TransformerMixin

class CleanJobClassification(BaseEstimator, TransformerMixin):
    def __init__(self):
        super().__init__()

    def fit(self, X , y=None):
        return self

    def transform(self,X,y=None):
        x_copy = pd.DataFrame(X.copy())
        regex_df = x_copy.apply(lambda x: x.str.extract(r'^.* - (.*$)', expand=False))
        x_copy = regex_df.combine_first(x_copy)
        return x_copy
```

```python
class BinOutliers(BaseEstimator, TransformerMixin):
    def __init__(self, upper_bound=100.0, lower_bound=0.0, bin_value=0):
        super().__init__()
        self.upper_bound = upper_bound
        self.lower_bound = lower_bound
        self.bin_value = bin_value

    def fit(self, X , y=None):
        return self

    def transform(self,X,y=None):
        x_copy = X.copy()
        x_copy = x_copy.apply(pd.to_numeric, errors='coerce')
        return pd.DataFrame([self.bin_value if val > self.upper_bound or val < self.lower_bound else val for val in x_copy.T])
```

```python
import time
from sklearn.base import BaseEstimator, TransformerMixin


class ThresholdBinner(BaseEstimator, TransformerMixin):
    def __init__(self, threshold=0.01, bin_label='other'):
        super().__init__()
        self.threshold = threshold
        self.bin_label = bin_label
        self.binned_categories = None

    def bin_categories(self, X, y=None):
        # Find the categories that make up a population smaller than the threshold
        x_copy = X.copy()
        self.binned_categories = []
        for col in x_copy.T:
            uniques, counts = np.unique(col.astype(str), return_counts=True)
            self.binned_categories.append(uniques[np.where(counts/len(col) <= self.threshold)])
        return

    def fit(self, X , y=None):
        # The fit function is for setting the parameters of the class
        t0 = time.time()
        self.bin_categories(X, y)
        t1 = time.time()
        print("total fit time: {:.2f} seconds".format(t1-t0))
        return self

    def apply_bin_label(self, vec, idx):
        t0 = time.time()
        vals_to_bin_idxs = (vec.astype(str)==self.binned_categories[idx].reshape(-1, 1)[:,None]).any(0)[0].nonzero()[0]
        vec[vals_to_bin_idxs] = self.bin_label
        t1 = time.time()
        print("apply bin label time: {:.2f} seconds".format(t1-t0))
        return vec

    def transform(self,X,y=None):
        # Apply the parameters of the class
        t0 = time.time()
        value = np.array([self.apply_bin_label(v,i) for i,v in enumerate(X.copy().T)]).T
        t1 = time.time()
        print("total transform time: {:.2f} seconds".format(t1-t0))
        return value
```
