## Command Line Execution
A python file is named with the extension .py\
`file.py` \
You can read and create a file in a text edit such as `vim` \
`vim file.py` \
To run (execute the file (it run the program, line by line, top to bottom) use `Python` \
`Python filename.py` \
The file is compiled and executed in one step

## Numeric Operations

```python
# You can use different 'operators' in python
# The + operator adds the nunbers to the left and the right of the operator
# This is integer addition where int + int = int
# float + int = flaot

my_var1 = 72 + 36
my_float = 1.3 + 35

# the - operator supbrats the numbers from left to right of the operator
my_var2 = 300 / 2

# division always returns a floating point number
my_var4 = 5 / 2
my_var5 = 100 / 3

# the * operator multiplies the number to the left and the right of the operator
my_var6 = 2 * 2
```

```python
# Remainder operator
remainder = 100 % 2
print(remainder)

print(3%2)
print(50%3)
```

```python
# Incrementing
# These statments mean exactly the same thing:
count = count + 1
count += 1

# We can increment a variable by any number we like.
count += 2
count += 7
count += a + b

# These statements mean exactly the same thing
count = count -3
count -= 3
```

## String Operations

```python
# Strings are lists of characters
# Make a 4 character string and assign it to a name
s = 'spam'
print(len(s))

# Indexing is zero based
# The index of the string at 0
print(s[0])

# the second item from the left
print(s[1])

# you can grab the end of the string
# right most item
print(s[-1])

# the second item from the right
print(s[-2])
```

```python
# slicing
# slicing prints from the first number given to the secon dnumber given, but not including the second number
# Slice of s from 1 to 3, but not including 3
print(s[1:3])

# first position (index) to last
print(s[1:])

# first item (zeroth position) to the third index (but not including the third index)
print(s[:3])
print(s[0:-1])

# everything
print(s[:])
```

```python
# concatination
print(s + 'xyz')

# Strings are immutable - they cannot be cahnged in place after they are created
# The actual variables is unchanged because we did not assign the change
print(s)

# repetition
print(s*8)
```

```python
# strings are immutable
s[0] = 'z'
```

```python
# However, remember that you can call strings at a position
s[0]
```

```python
# instead, we would need to create a new object
s = 'z' + s[1:]
print(s)
```

```python
# global variable assignemnt
a = 0

# local use of a global variable
def my_function():
    a = 3
    print(a)
    
# call your function
my_function()

# call your global variable
print(a)

# Results: your local function can use your globabl variable but your globabl variable isn't change by your local function
```
